```python
import pandas as pd
import requests
import os

# Ruta del archivo
archivo_excel = r'C:\Users\gaas\OneDrive - SERVIFOR, S.C\Personal\Proyectos_personales\NFL\2025\marcador_2025.xlsx'



tipo_temporada = 2
temporada_regular = {}

for semana in range(1, 19):
    url = f"https://site.api.espn.com/apis/site/v2/sports/football/nfl/scoreboard?week={semana}&seasontype={tipo_temporada}"
    response = requests.get(url)
    data = response.json()

    partidos = []
    for event in data.get("events", []):
        nombre_partido = event.get("name")
        fecha = event.get("date")
        competencia = event.get("competitions", [])[0]
        equipos = [equipo["team"]["name"] for equipo in competencia.get("competitors", [])]
        equipo_1 = nombre_partido.split(" at ")[0] if " at " in nombre_partido else None
        equipo_1 = equipos_nfl.get(equipo_1, equipo_1)
        equipo_2 = nombre_partido.split(" at ")[1] if " at " in nombre_partido else None
        equipo_2 = equipos_nfl.get(equipo_2, equipo_2)
        estadio = competencia.get("venue", {}).get("fullName", "Desconocido")
        puntos_equipo_1 = competencia.get("competitors", [])[1].get("score", 0)
        puntos_equipo_2 = competencia.get("competitors", [])[0].get("score", 0)

        partidos.append({
            "Partido": nombre_partido,
            "Fecha": fecha,
            "Equipo 1": equipo_1,
            "Puntos Equipo 1": puntos_equipo_1,
            "Equipo 2": equipo_2,
            "Puntos Equipo 2": puntos_equipo_2,
            "Estadio": estadio,
        })

    df = pd.DataFrame(partidos)
    df['combinación'] = df['Estadio'] + " " + df['Equipo 2']
    df['Local'] = df['combinación'].map(estadios_nfl)
    df['Local'] = df['Local'].fillna(df['Equipo 2'])
    df['Ganador'] = df.apply(lambda row: row['Equipo 1'] if int(row['Puntos Equipo 1']) > int(row['Puntos Equipo 2']) else (row['Local'] if int(row['Puntos Equipo 2']) > int(row['Puntos Equipo 1']) else 'Empate'), axis=1)

    temporada_regular[semana] = df

# Guardar todas las semanas en el archivo Excel
with pd.ExcelWriter(archivo_excel, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    for semana, df in temporada_regular.items():
        df.to_excel(writer, sheet_name=str(semana), index=False)



# Ruta del archivo
archivo = r'C:\Users\gaas\OneDrive - SERVIFOR, S.C\Personal\Proyectos_personales\NFL\2025\marcador_2025.xlsx'

# Diccionario para almacenar ganadores por semana
ganadores_por_semana = {}

for semana in range(1, 19):
    nombre_hoja = str(semana)
    try:
        df = pd.read_excel(archivo, sheet_name=nombre_hoja, engine='openpyxl')
        if 'Ganador' in df.columns:
            ganadores_por_semana[semana] = df['Ganador'].tolist()
        else:
            ganadores_por_semana[semana] = ['Columna "Ganador" no encontrada']
    except Exception as e:
        ganadores_por_semana[semana] = ['Hoja no encontrada']

# Crear DataFrame con índice como semana
df_ganadores_semanales = pd.DataFrame.from_dict(ganadores_por_semana, orient='index')
df_ganadores_semanales.index.name = 'Semana'

# Mostrar el DataFrame
df_ganadores_semanales
```
