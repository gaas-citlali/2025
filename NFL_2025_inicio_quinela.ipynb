```python
import pandas as pd
import os
import warnings


archivo = f'Semana {numero} NFL 2025.xlsx'

fecha_inicio = pd.to_datetime('2025-09-01')

warnings.filterwarnings("ignore")
respuestas = pd.read_excel(archivo)

respuestas['fecha'] = respuestas['Completion time'].dt.date
respuestas['semana'] = (respuestas['Completion time'] - fecha_inicio).dt.days // 7 + 1

columnas = ['semana', 'fecha']
columnas_viejas = [col for col in respuestas.columns if col not in columnas]

respuestas = respuestas[columnas + columnas_viejas]
respuestas = respuestas.rename(columns={'Name':'Participante'})

correos = respuestas['Email'].unique()

respuestas
```
