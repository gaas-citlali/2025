# Definir aciertos semanales

```python
def contar_aciertos(quiniela, resultados):
    import pandas as pd

    def limpiar_nombre(equipo):
        if isinstance(equipo, str):
            return equipo.replace("(Local)", "").strip().lower()
        return equipo
    acierto_semanal = pd.DataFrame()

    for semana, quiniela_semana_actual in quiniela.items():
        aciertos_semana = {}

        try:
            semana_int = int(semana)
        except ValueError:
            print(f"Semana '{semana}' no es un n√∫mero v√°lido.")
            continue

        if semana_int not in resultados.index:
            print(f"Semana {semana_int} no encontrada en resultados")
            continue

        resultado_real = resultados.loc[semana_int]
        resultado_real = resultado_real.apply(limpiar_nombre)

        for participante in quiniela_semana_actual.index:
            seleccion_participante = quiniela_semana_actual.loc[participante]
            seleccion_participante = seleccion_participante.reindex(resultado_real.index)
            seleccion_participante = seleccion_participante.apply(limpiar_nombre)

            aciertos_semana[participante] = (seleccion_participante == resultado_real).sum()

        acierto_semanal[semana_int] = pd.Series(aciertos_semana)

    acierto_semanal.fillna(0, inplace=True)
    acierto_semanal = acierto_semanal.astype(int)
    acierto_semanal['Total Aciertos'] = acierto_semanal.sum(axis=1)

    return acierto_semanal
```
