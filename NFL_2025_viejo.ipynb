```python
def contabilizar_aciertos_semanales(quinela_dict, resultados_df):
    aciertos_semanales_df = pd.DataFrame()

    for semana, quinela_df in quinela_dict.items():
        aciertos_semana = {}
        
        for participante in quinela_df.index:
            seleccion_participante = quinela_df.loc[participante]
            resultado_real = resultados_df.loc[int(semana)]

            # Reindexar para asegurarse de que los índices coincidan
            seleccion_participante = seleccion_participante.reindex(resultado_real.index)

            # Contar los aciertos
            aciertos_semana[participante] = (seleccion_participante == resultado_real).sum()

        aciertos_semanales_df[semana] = pd.Series(aciertos_semana)

    aciertos_semanales_df.fillna(0, inplace=True)
    aciertos_semanales_df = aciertos_semanales_df.astype(int)
    aciertos_semanales_df['Total Aciertos'] = aciertos_semanales_df.sum(axis=1)

    return aciertos_semanales_df
```
